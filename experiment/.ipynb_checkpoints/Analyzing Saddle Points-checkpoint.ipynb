```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random
```

Below is a plot for function z = x^2 - y^2 <br/>
<b>Saddle Point</b> = (0, 0) <br/>
<b>Function Minima</b> = (0, n)

```python
def saddleFunc(x, y):
  return x**2 - y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-4.0, 4.0, 0.01)
X, Y = np.meshgrid(x, y)
zs = np.array([saddleFunc(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('Weight1 - X')
ax.set_ylabel('Weight2 - Y')
ax.set_zlabel('Loss - Z')

plt.show()
```

Below we are not using any momentum(concept we use to help take us out of saddle points). We see that we are always stuck at the saddle point. Here weights were deliberately initialized to (0, 0) - Saddle Point. We can expect that during learning process if weights end up at saddle point then it would be difficult for Gradient Descent to come out.

```python
epochs = 15
X_ini = 0
Y_ini = 0
learn_rate = 0.1
momentum = 0.0
loss_ini = X_ini**2 - Y_ini**2

for epoch in range(epochs):
    X_ini -= learn_rate*2*X_ini + momentum
    Y_ini -= learn_rate* (-2*Y_ini) + momentum
    loss_ini = X_ini**2 - Y_ini**2
    print("Loss = " + str(loss_ini) + ", X = " + str(X_ini) + ", Y = " + str(Y_ini))
```

Below we are using a small momentum(Analogy from Physics). We do come out of the saddle point and the learning occurs.

```python
epochs = 15
X_ini = 0
Y_ini = 0
learn_rate = 0.1
momentum = 0.01
loss_ini = X_ini**2 - Y_ini**2

for epoch in range(epochs):
    X_ini -= learn_rate*2*X_ini + momentum
    Y_ini -= learn_rate* (-2*Y_ini) + momentum
    loss_ini = X_ini**2 - Y_ini**2
    print("Loss = " + str(loss_ini) + ", X = " + str(X_ini) + ", Y = " + str(Y_ini))
```
